### BASIC EXAMPLE USAGE OF DATASIST
#### We use the iris data set for this sample analysis

```python
import pandas as pd
import datasist as ds #import datasist library

df = pd.read_csv('iris.csv')
```

**Structdata** modules in datasist contains functions used to manipulating structured data in pandas or csv format.

```python
ds.structdata.describe(df)
```

```python
#Gets all categorical features
cats = ds.structdata.get_cat_feats(df)
cats
```

```python
#Gets all numeric features
nums = ds.structdata.get_num_feats(df)
nums
```

```python
#Gets the unique count of all categorical features
ds.structdata.get_unique_counts(df)
```

```python
#Display missing values in data set
ds.structdata.display_missing(df)
```

#### VISUALIZATION
**visualization** module in datasist contains abstracted functions for making popular EDA plots without writing complex for loops

```python
ds.visualizations.countplot(df)
```

```python
ds.visualizations.boxplot(data=df, target='species')
```

```python
ds.visualizations.histogram(df)
```

```python
ds.visualizations.violinplot(data=df, target='species')
```

```python
ds.visualizations.scatterplot(data=df, target='petal_width',separate_by='species', fig_size=(6,6))
```
